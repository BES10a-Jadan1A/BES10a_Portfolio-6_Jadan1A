```python
# =========================================================
# KEMENG LABORATORY INVENTORY LIST SYSTEM
# =========================================================
# This program uses Python LISTS to manage laboratory
# chemicals commonly used by Chemical Engineering students.

# This simple system helps students monitor chemical
# inventory, concentrations, and laboratory usage.
#
# Significance:
# Proper chemical inventory management helps improve
# laboratory safety, organization, and experiment accuracy.
# =========================================================
```

```python
# Empty lists for laboratory records

chemical_names = []
chemical_concentrations = []
chemical_volumes = []
```

```python
print("KEMENG INVENTORY SYSTEM")
print("Type 'done' to stop.\n")

while True:

    chemical = input(
        "Enter chemical name: "
    )

    if chemical.lower() == "done":
        break

    concentration = float(input(
        "Enter concentration (mol/L): "
    ))

    volume = float(input(
        "Enter volume (mL): "
    ))

    # append() method
    chemical_names.append(chemical)
    chemical_concentrations.append(concentration)
    chemical_volumes.append(volume)

    print("Chemical successfully added.\n")
```

```python
print("\n========== KEMENG CHEMICAL RECORDS ==========")

# Loop through lists
for i in range(len(chemical_names)):

    print("\nChemical #", i + 1)

    # List indexing
    print("Name:",
          chemical_names[i])

    print("Concentration:",
          chemical_concentrations[i],
          "mol/L")

    print("Volume:",
          chemical_volumes[i],
          "mL")
```

```python
# Built-in list functions

total_chemicals = len(chemical_names)

highest_concentration = max(
    chemical_concentrations
)

lowest_concentration = min(
    chemical_concentrations
)

total_volume = sum(
    chemical_volumes
)

average_volume = (
    total_volume / total_chemicals
)
```

```python
# Sort chemical names alphabetically

chemical_names.sort()

print("\nSorted Chemical Names:")
print(chemical_names)

# List slicing
print("\nFirst Two Chemicals:")

print(chemical_names[:2])
```

```python
# Membership checking

search_chemical = input(
    "\nEnter chemical to search: "
)

if search_chemical in chemical_names:

    print(search_chemical,
          "is available in inventory.")

else:

    print(search_chemical,
          "is NOT available.")
```

```python
# split() demonstration

chemical_data = (
    "Ammonium thioglycolate;Thiocyanate;Acetic acid;Ammonium hydroxide"
)

chemical_list = chemical_data.split(";")

print("\nFinal KEMENG Chemical List:")
print(chemical_list)
```

```python
print("\n========== KEMENG LABORATORY SUMMARY ==========")

print("Total Chemicals:",
      total_chemicals)

print("Highest Concentration:",
      highest_concentration,
      "mol/L")

print("Lowest Concentration:",
      lowest_concentration,
      "mol/L")

print("Total Volume Used:",
      total_volume,
      "mL")

print("Average Volume:",
      round(average_volume, 2),
      "mL")

print("\nLaboratory inventory updated successfully.")
```
